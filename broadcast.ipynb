```python
import numpy as np
import matplotlib.pyplot as plt
import time

num_x_vals = 100
x_initial_values = [i/(num_x_vals-1) for i in range(num_x_vals)]

num_r_vals = 1000
r_vals = [i*4/(num_r_vals-1) for i in range(num_r_vals)]

num_timesteps = 600
```

```python
start_time = time.time()

x_initial_values = np.array(x_initial_values)
r_vals = np.array(r_vals)

# x_final_vs_r = x_initial_values[:, np.newaxis]*(1-x_initial_values[:, np.newaxis])*r_vals
x_final_vs_r = (x_initial_values*(1-x_initial_values))[:, np.newaxis]*r_vals
for _ in range(num_timesteps-1):
    x_final_vs_r = x_final_vs_r*(1-x_final_vs_r)*r_vals
    
print("Execution time: ", time.time()-start_time)
```

```python
plt.figure(figsize=(18,8))

for row in x_final_vs_r:
    _ = plt.plot(r_vals, row, c='red', linewidth=0.02)
    
plt.xlabel('r')
plt.ylabel('x')
```
