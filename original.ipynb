```python
import matplotlib.pyplot as plt
import time

num_x_vals = 100
x_initial_values = [i/(num_x_vals-1) for i in range(num_x_vals)]

num_r_vals = 1000
r_vals = [i*4/(num_r_vals-1) for i in range(num_r_vals)]

num_timesteps = 600
```

```python
start_time = time.time()
xr_matrix = []

for i in range(num_x_vals):
    x_final_vs_r = []
    for r in r_vals:
        x = x_initial_values[i]
        for _ in range(num_timesteps):
            x = x * (1-x) * r
        x_final_vs_r.append(x)
    xr_matrix.append(x_final_vs_r)
    
print("Execution time: ", time.time()-start_time)
```

```python
plt.figure(figsize=(18,8))

for row in xr_matrix:
    _ = plt.plot(row, c='red', linewidth=0.02)
```
